```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# We'll create synthetic data
scale = 5
size = 500
```

```python
# First and second normal distribution:
class_1_x = np.random.normal(loc = 10, scale=scale, size=size)
class_1_y = np.random.normal(loc = 1, scale=scale, size=size)
```

```python
# First and second normal distribution:
class_2_x = np.random.normal(loc = 1, scale=scale, size=size)
class_2_y = np.random.normal(loc = 5, scale=scale, size=size)
```

```python
# Plot the distributions above:
plt.figure(figsize=(7,7))
plt.scatter(class_1_x, class_1_y)
plt.scatter(class_2_x, class_2_y)
plt.show()
```

```python
df_class1 = pd.DataFrame()
df_class1['x'] = class_1_x
df_class1['y'] = class_1_y
df_class1['class'] = 0
```

```python
df_class2 = pd.DataFrame()
df_class2['x'] = class_2_x
df_class2['y'] = class_2_y
df_class2['class'] = 1
```

```python
df = pd.concat([df_class1, df_class2], ignore_index=True)
```

```python
# Get the dependent and independent variables
X = df.drop(columns = ['class'])
y = df['class']
```

```python
# Split data to train:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Create the model object
model = GaussianNB()
```

```python
# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
# Predict the classes on the test data
y_pred = model.predict(X_test)
```

```python
# Show the first 10 entries
y_pred[:10]
```

```python
# Predict the classes on the test data, and return the probabilities for each class
y_proba = model.predict_proba(X_test)
```

```python
y_proba[:10]
```

```python
import seaborn as sns
```

```python
# Show the predictions
sns.scatterplot(x = X_test['x'], y = X_test['y'], hue = y_pred)
```

## Plot the ROC Curve

```python
from sklearn.metrics import confusion_matrix
```

```python
def calculate_tpr_fpr(y_real, y_pred):
    '''
    Calculates the True Positive Rate (tpr) and the True Negative Rate (fpr) based on real and predicted observations
    
    Args:
        y_real: The list or series with the real classes
        y_pred: The list or series with the predicted classes
        
    Returns:
        tpr: The True Positive Rate of the classifier
        fpr: The False Positive Rate of the classifier
    '''
    
    # Calculates the confusion matrix and recover each element
    cm = confusion_matrix(y_real, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    
    # Calculates tpr and fpr
    tpr =  TP/(TP + FN) # sensitivity - true positive rate
    fpr = 1 - TN/(TN+FP) # 1-specificity - false positive rate
    
    return tpr, fpr
```

```python

def get_n_roc_coordinates(y_real, y_proba, resolution = 50):
    '''
    Calculates "n" ROC Curve coordinates (tpr and fpr) by manipulating the threshold used to predict the class.
    
    Args:
        y_real: The list or series with the real classes.
        y_proba: The array with the probabilities for each class, obtained by using the `.predict_proba()` method.
        resolution: Defines how many divisions the threshold will have, and how many coordinates will be calculated (default = 50).
        
    Returns:
        tpr_list: The list of TPRs representing each threshold.
        fpr_list: The list of FPRs representing each threshold.
    '''
    tpr_list = [0]
    fpr_list = [0]
    for i in range(resolution):
        threshold = i/resolution
        y_pred = y_proba[:, 1] > threshold
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

```python
def plot_roc_curve(tpr, fpr, scatter = True):
    '''
    Plots the ROC Curve by using the list of coordinates (tpr and fpr).
    
    Args:
        tpr: The list of TPRs representing each coordinate.
        fpr: The list of FPRs representing each coordinate.
        scatter: When True, the points used on the calculation will be plotted with the line (default = True).
    '''
    plt.figure(figsize = (5, 5))
    if scatter:
        sns.scatterplot(x = fpr, y = tpr)
    sns.lineplot(x = fpr, y = tpr)
    sns.lineplot(x = [0, 1], y = [0, 1], color = 'green')
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
```

```python
# Calculates 10 coordinates of the ROC Curve
tpr, fpr = get_n_roc_coordinates(y_test, y_proba, resolution = 10)
```

```python
# Plots the ROC curve
plot_roc_curve(tpr, fpr)
```

## Better version

Use all data points to calculate the coordinates, and plots the ROC Curve

```python
def get_all_roc_coordinates(y_real, y_proba):
    '''
    Calculates all the ROC Curve coordinates (tpr and fpr) by considering each point as a treshold for the predicion of the class.
    
    Args:
        y_real: The list or series with the real classes.
        y_proba: The array with the probabilities for each class, obtained by using the `.predict_proba()` method.
        
    Returns:
        tpr_list: The list of TPRs representing each threshold.
        fpr_list: The list of FPRs representing each threshold.
    '''
    tpr_list = [0]
    fpr_list = [0]
    for i in range(len(y_proba)):
        threshold = y_proba[i, 1]
        y_pred = y_proba[:, 1] >= threshold
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

```python
# Calculates ALL coordinates of the ROC Curve
tpr, fpr = get_all_roc_coordinates(y_test, y_proba)
```

```python
# Plots the ROC curve
plot_roc_curve(tpr, fpr, scatter = False)
```

### Compare with Sklearn


```python
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
```

```python
def plot_sklearn_roc_curve(y_real, y_pred):
    '''
    Plots the ROC Curve with the sklearn methods by using the real observations and their predictions.
    
    Args:
        y_real: The list or series with the real classes
        y_pred: The list or series with the predicted classes
    '''
    fpr, tpr, _ = roc_curve(y_real, y_pred)
    roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
    roc_display.figure_.set_size_inches(5,5)
```

```python
# Plots the ROC curve using the sklearn methods
plot_sklearn_roc_curve(y_test, y_proba[:, 1])
```
