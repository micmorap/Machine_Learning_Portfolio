**El descenso del gradiente se utiliza para alcanzar el mÃ­nimo de la funciÃ³n de error(se produce al dar los valores iniciales a los pesos de la neuronas de la red neuronal y el bias), para actualizar los pesos y el bias o sesgo en el entrenamiento de la red neuronal profunda artificial**

```python
# Importamos los mÃ³dulos a usar.
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creamos la funciÃ³n a optimizar y la grafica.
# lambda es un tipo de funcion llamada anonima.

funcion = lambda x: x**2+ 1
# x**2+ 5*x+ 6
# se crean 200 valores espaciados entre x=-10 y x=10
x = np.linspace(-10,10,200)
# Grafica la funciÃ³n.
plt.plot(x, funcion(x))
plt.show()
```

```python
# Creamos la funciÃ³n de la derivada de x.
# utilizamos def nombredelafuncion():
# return=valor de retorno: resultado
def derivada(x):

    x_derivada = 2*x 
    return x_derivada # Regresa la derivada de la funciÃ³n inicial

# Creamos la funciÃ³n del gradiente
def gradiente(x_inicio, precision, l_r):

# se crean dos listas para los valores de x,y 
#se utilizaran despues para la visualizacion puntual.
   
    

    x_list, y_list = [x_inicio], [funcion(x_inicio)]

    # Utilizamos el bucle While para hacer el recorrido

    while True:

        ''' 
         Se Obtiene el valor de la pendiente de la derivada
        debido a que necesitamos un descenso negativo (hacia el mÃ­nimo)
        usamos el signo menos '-'

        '''

        d_x = - derivada(x_inicio)
        
        # Obtenemos el nuevo valor de x sumando el valor previo de x mÃ¡s la multiplicaciÃ³n del l_r por la derivada.
        x_inicio = x_inicio + (l_r * d_x)
        
        # Agregamos  o aÃ±adimos el nuevo valor de x a la lista para posteriormente visualizarlo.
        # metodo append() para agregar un valor a la lista
        x_list.append(x_inicio)
        
        # Agregamos o aÃ±adimos el nuevo valor de y a la lista para posteriormente visualizarlo.
        y_list.append(funcion(x_inicio))

       
        #abs significa valor absoluto
        # x_list[-1] significa Ãºltima valor en la lista
        # x_list[-2] significa valor antes del Ãºltimo
        # abs(x_list[-1] - x_list[-2]) <= precision. 
        #significa si la resta de los dos es menor que la precision pare.
        # Salimos del bucle cuando se cumpla la condiciÃ³n.
        
        if abs(x_list[-1] - x_list[-2]) <= precision:
            break

    # Visualizamos los resultados.  
    # str() convierte a string.
    # len() longitud de la lista.

    print ("El mÃ­nimo local se produce en: "+ str(x_inicio))
    
    # NÃºmero de pasos para obtener el minimo local.
    
    print ("NÃºmero de pasos: " + str(len(x_list)))
    
    # subplot para ver varias graficas en una.
    # visualizar la grafico de dispersion mediante plt.scatter()
    
    plt.subplot(1,2,2)# tambien se puede escribir plt.subplot(122)
    plt.scatter(x_list,y_list,c="b")# color=c="b" significa b=blue color azul,y= yellow(amarillo) 
    plt.plot(x_list,y_list,c="b")# grÃ¡fico para ver el recorrido que realizo el gradiente para obtener el minimo.
    plt.plot(x,funcion(x), c="y")# grÃ¡fico de la funcion inicial
    plt.title("GrÃ¡fico del descenso del gradiente paso a paso.")
    plt.show()
```

**LLamamos la funciÃ³n gradiente pasando los parametros, punto inicial de x,precision y learning rate(tasa de aprendizaje)**

```python
gradiente(-9, 0.0001, 0.01)
```

**Varien esos 3 valores del Descenso del Gradiente**

```python
gradiente(-8, 0.01, 0.03)
```

```python
gradiente(5, 0.01, 0.03)
```
