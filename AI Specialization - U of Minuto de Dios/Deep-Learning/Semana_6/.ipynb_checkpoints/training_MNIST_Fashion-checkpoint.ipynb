### Objective:

The aim of this notebook is replicate the training exercise exposed in the Tensorflow webpage (https://www.tensorflow.org/tutorials/keras/classification?hl=es-419) which train a neural network to workt with MNIST fashion dataset.

All the descriptions and explanations about this dataset could be find in the url above.

### Imports and EDA:


```python
# Import basic packages and libraries:
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# validate our tensorflow current version
print(tf.__version__)
```

```python
# Import dataset from keras:
fashion_mnist = keras.datasets.fashion_mnist
```

```python
# Split the data into training and testing set:
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Review the length of our samples:
print("Number images to train: {}".format(train_images.shape[0]))
print("Number labels to train: {}".format(train_labels.shape[0]))
print("Number images to test: {}".format(test_images.shape[0]))
print("Number labels to test: {}".format(test_labels.shape[0]))
```

Now, according to the official Tensorflow webpages, the labels going on from zero until nine with the next descriptions:

* Label - Class
* 0: T-shirt/top
* 1: Trouser
* 2: Pullover
* 3: Dress
* 4: Coat
* 5: Sandal
* 6: Shirt
* 7: Sneaker
* 8: Bag
* 9: Ankle boot

By suggestion of webpage, I'll include this categories in a list:

```python
# List of categories to each label:
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Review the shapes of our samples:
print("Dimensions to train: {}".format(train_images.shape))
print("Dimensions labels to train: {}".format(train_labels.shape))
print("Dimensions images to test: {}".format(test_images.shape))
print("Dimensions labels to test: {}".format(test_labels.shape))
```

### Preprocessing:

We visualize one randomly picture from the training set:


```python
plt.figure()
plt.imshow(train_images[90])
plt.colorbar()
plt.grid(False)
plt.show()
```

Now, We'll normalize our imagen from 0 to 1 (currently, as you look in the picture above, the ranges values going from 0 to 255), because help us to create an algorith more efficient and save computational resources and convergence as well:test_labels

```python
train_images = train_images / 255
test_images = test_images / 255
```

Let's find out what are the unique values to our labels:

```python
unique_labels = set(test_labels)
print('The unique labels are: {}'.format(unique_labels))
```

To make sure that the images has been normalized correctly, We'll visualize any pictures:

```python
plt.figure(figsize=(10, 10))
for i in range(20):
    plt.subplot(5, 4, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
```

### Building and compiling the model

Now, We'll create the layers and then instantiate the respective model:

```python
# Create the model:
model = keras.Sequential([
        keras.layers.Flatten(input_shape=(28,28)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(10, activation='softmax')  
        ])
```

```python
# Compiling the model:
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
```

### Training the model:

Then, We fit the model indicates the train set (images and labels) and the number epochs:

```python
# train the model
model.fit(train_images, train_labels, epochs=20)
```

Our model got a 94% of accuracy with a loss of 14.12% until the last execution (epoch) with training set.

Now, We'll evaluate the accuracy with test set:

```python
test_loss, test_accuracy = model.evaluate(test_images, test_labels, verbose=2)

print('Test accuracy:', test_accuracy)
```

Our model got a 88% of accuracy with a loss of 42% with testing set

### Make predictions

Finally, We make predictions about a couple of images and visualize the results:

```python
# predict function to test images:
predictions = model.predict(test_images)
```

```python
# We take a random picture:
predictions[36]
```

```python
# Review the high value of the last array that contains 10 probabilities values:
np.argmax(predictions[36])
```

The last value (seven) means that our model predicted that the fashion objects would be a sneaker. What is the real result?

```python
# reviewing the real value
```

```python
print('The real label of the image number 36 is: {}, a {}'.format(test_labels[36], class_names[test_labels[36]]))
```

```python
# We'll reply two functions to graph this results:
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

```python
# Review the 77th image:
i = 77
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

```python
# Plot the first X test images, their predicted labels, and the true labels.
# Color correct predictions in blue and incorrect predictions in red.
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions[i], test_labels)
plt.tight_layout()
plt.show()
```
