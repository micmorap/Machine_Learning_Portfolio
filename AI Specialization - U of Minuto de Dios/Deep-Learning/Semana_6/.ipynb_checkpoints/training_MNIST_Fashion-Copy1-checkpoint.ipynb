### Imports and EDA:


```python
# Import basic packages and libraries:
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
```

```python
# validate our tensorflow current version
print(tf.__version__)
```

#### Replicate any special functions:

```python
def preprocess(array):
    """
    Normalizes an array (divided in 255) to get pixel values between 0 and 1
    and reshape in the suitable way.
    """
    
    array = array.astype("float32") / 255
    array = np.reshape(array, (len(array), 28, 28, 1))
    return array
    
```

```python
def noise(array):
    """
    Adds random noise to each image in the supplied array.
    """

    noise_factor = 0.6
    noisy_array = array + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=array.shape)

    return np.clip(noisy_array, 0.0, 1.0)
```

```python
def display(array1, array2):
    """
    Displays ten random images from each one of the supplied arrays.
    """

    n = 10

    indices = np.random.randint(len(array1), size=n)
    images1 = array1[indices, :]
    images2 = array2[indices, :]

    plt.figure(figsize=(20, 4))
    for i, (image1, image2) in enumerate(zip(images1, images2)):
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(image1.reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(image2.reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

    plt.show()
```

#### Prepare the data:

```python
# We don't need the labels info from MNIST dataset:
(train_data,_), (test_data, _) = mnist.load_data()
```

Now, We'll normalize our imagen from 0 to 1 (currently, as you look in the picture above, the ranges values going from 0 to 255), because help us to create an algorith more efficient and save computational resources and convergence as well:test_labels

```python
train_data = preprocess(train_data)
test_data = preprocess(test_data) 
```

```python
# Review the length of our samples:
print("Number images to train: {}".format(train_data.shape[0]))
print("Number image to test: {}".format(test_data.shape[0]))
print("Shape train set: {}".format(train_data.shape))
print("Shape test set: {}".format(test_data.shape))
```

Now, We'll create a noisy dataset with **noise** function and compare the original and noisy images:


```python
noisy_train_data = noise(train_data)
noisy_test_data = noise(test_data)
```

```python
display(train_data, noisy_train_data)
```

### Building the autoencoder:

```python
input = layers.Input(shape=(28,28,1))
```

```python
# Encoder
# Encoder
x = layers.Conv2D(32, (3, 3), activation="relu", padding="same")(input)
x = layers.MaxPooling2D((2, 2), padding="same")(x)
```

```python
# Decoder
x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding="same")(x)
x = layers.Conv2D(1, (3, 3), activation="sigmoid", padding="same")(x)
```

```python
# Autoencoder
autoencoder = Model(input, x)
autoencoder.compile(optimizer="adam", loss="binary_crossentropy")
autoencoder.summary()
```

#### Training the model:

```python
autoencoder.fit(
    x=train_data,
    y=train_data,
    epochs=50,
    batch_size=128,
    shuffle=True,
    validation_data=(test_data, test_data),
)
```

Let's find out what are the unique values to our labels:

#### Predict results with the model trained:

In general, We must to see a best performance of the pictures after the autoencoder was applied:

```python
predictions = autoencoder.predict(test_data)
display(test_data, predictions)
```

Now, We'll replay the fit step to train the autoencoder but with noisy data:

```python
autoencoder.fit(
    x = noisy_train_data,
    y = train_data,
    epochs = 50,
    batch_size = 128,
    shuffle= True,
    validation_data=(noisy_test_data, test_data),    
)
```

```python
predictions = autoencoder.predict(noisy_test_data)
display(noisy_test_data, predictions)
```
