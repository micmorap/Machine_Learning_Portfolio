```python
# Import packages:
import cv2
import skimage
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import canny
from skimage import io, feature
from skimage.filters import try_all_threshold
from skimage.filters import threshold_mean
```

#### Definimos una función que itere a lo largo de cada filtro sobre las 4 imágenes:

```python
def iterador_images(listado: list, tipo_filtro: str):
    # Crear una figura y subplots en una fila
    fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(8, 5))
    
    for a in ax.flat:
        a.axis('off')    
    
    for m in range(4):
        ax[m].imshow(listado[m])
        ax[m].set_title("Filtro {}".format(tipo_filtro))
    
    plt.tight_layout()
```

```python
def iterador_images_2(listado: list, tipo_filtro: str):
    # Crear una figura y subplots en una fila
    fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(8, 5))
    
    for a in ax.flat:
        a.axis('off')    
    
    for m in range(4):
        ax[m].imshow(listado[m], cmap='gray')
        ax[m].set_title("Filtro {}".format(tipo_filtro))
        
    
    plt.tight_layout()
```

#### Definimos 8 listas con las imágenes resultantes para cada filtro:

```python
nombre = "imagen_"
original_images_list = []
for i in range(1,5):
    imagen = cv2.imread("pictures/image_" + str(i) + ".jpeg")
    original_images_list.append(imagen)
```

```python
def show_image_color(input_image: list):
    """
    Converts and show an image in gray scale
    """
    #Change color channels order for matplotlib  
    list_original_images = []
    
    for i in range(len(input_image)):
        color_output = cv2.cvtColor(input_image[i], cv2.COLOR_BGR2RGB) 
        list_original_images.append(color_output)
    
    return list_original_images
```

```python
# Función para escala de grises:
def show_image_gray(input_image_list: list):
    """
    Converts and show an image in gray scale
    """
    #Change color channels order for matplotlib    
    list_gray_images = []
    
    for i in range(len(input_image_list)):
        gray_output = cv2.cvtColor(input_image_list[i], cv2.COLOR_BGR2GRAY) 
        list_gray_images.append(gray_output)
        
    return list_gray_images
```

```python
# Función para calcular el filtro Laplaciano:

def show_laplacian_filter(input_image_list: list):
    """
    Converts and show an image with the Sobel filter applied
    """
    #Change color channels order for matplotlib    
    list_laplacian_images = []
    
    for i in range(len(input_image_list)):
        lapl_output = cv2.Laplacian(input_image_list[i],cv2.CV_64F, ksize=5)
        list_laplacian_images.append(lapl_output)
    
    return list_laplacian_images

        
```

```python
# Función para calcular el filtro Sobel:

def show_sobel_filter(input_image_list: list):
    """
    Converts and show an image with the Sobel filter applied
    """
    #Change color channels order for matplotlib    
    list_sobel_images = []
    
    for i in range(len(input_image_list)):
        sobel_filter_x = cv2.Sobel(input_image_list[i], cv2.CV_64F, 1, 0, ksize=5)
        sobel_filter_y = cv2.Sobel(input_image_list[i], cv2.CV_64F, 0, 1, ksize=5)
        list_sobel_images.append(sobel_filter_y)
    
    return list_sobel_images

        
```

```python
# Función para calcular el filtro Gaussian blurred:

def show_gaussian_blurred_filter(input_image_list: list):
    """
    Converts and show an image with the Gaussian blurred filter applied
    """
    #Change color channels order for matplotlib    
    list_gaussian_blurred_images = []
    kernel_size = (5,5) 
    
    for i in range(len(input_image_list)):
        gaussian_burred_filter = cv2.GaussianBlur(input_image_list[i], kernel_size,0)
        list_gaussian_blurred_images.append(gaussian_burred_filter)
    
    return list_gaussian_blurred_images

        
```

```python
# Función para calcular el filtro Histogram equalized 1:

def show_histogram_filter(input_image_list: list):
    """
    Converts and show an image with the Gaussian blurred filter applied
    """
    #Change color channels order for matplotlib    
    list_histogram_images = []
    
    for i in range(len(input_image_list)):
        histogram_filter = cv2.equalizeHist(input_image_list[i])
        list_histogram_images.append(histogram_filter)
    
    return list_histogram_images
```

```python
# Función para calcular el filtro Histogram CLAHE:

def show_histogram_filter(input_image_list: list):
    """
    Converts and show an image with the Gaussian blurred filter applied
    """
    #Change color channels order for matplotlib    
    list_histogram_images = []
    
    for i in range(len(input_image_list)):
        histogram_filter = cv2.createCLAHE().apply(input_image_list[i])
        list_histogram_images.append(histogram_filter)
    
    return list_histogram_images
```

```python
# Función para calcular el filtro Canny:

def show_canny_filter(input_image_list: list):
    """
    Converts and show an image with the Gaussian blurred filter applied
    """
    #Change color channels order for matplotlib    
    list_canny_images = []
    
    for i in range(len(input_image_list)):
        canny_filter = feature.canny(input_image_list[i], sigma=1)
        list_canny_images.append(canny_filter)
    
    return list_canny_images
```

```python
# Función para calcular el filtro Threshold:

def show_threshold_filter(input_image_list: list):
    """
    Converts and show an image with the Gaussian blurred filter applied
    """
    #Change color channels order for matplotlib    
    list_threshold_images = []
    
    for i in range(len(input_image_list)):
        threshold_filter = threshold_mean(input_image_list[i])
        binary = input_image_list[i] > threshold_filter
        list_threshold_images.append(binary)
    
    return list_threshold_images
```

```python
# Lista
color_images_list = show_image_color(original_images_list)
gray_images_list = show_image_gray(color_images_list)
laplacian_images_list = show_laplacian_filter(gray_images_list)
sobel_images_list = show_sobel_filter(gray_images_list)
gaussian_blurred_list = show_gaussian_blurred_filter(original_images_list)
histogram_equalized_list = show_histogram_filter(gray_images_list)
clahe_list = show_histogram_filter(gray_images_list)
canny_list = show_canny_filter(gray_images_list)
threshold_list = show_threshold_filter(gray_images_list)
```

```python
iterador_images(color_images_list, "Original")
iterador_images_2(gray_images_list, "Escala grises")
iterador_images_2(laplacian_images_list, "Laplaciano")
iterador_images_2(sobel_images_list, "Filtro Sobel eje Y")
iterador_images_2(gaussian_blurred_list, "Gaussian blurred")
iterador_images_2(histogram_equalized_list, "Histogram equ.")
iterador_images_2(clahe_list, "CLAHE")
iterador_images_2(canny_list, "Canny")
iterador_images_2(threshold_list, "Threshold")
```
